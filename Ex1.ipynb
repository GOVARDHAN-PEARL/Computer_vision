```python
import cv2
import matplotlib.pyplot as plt
```

```python
img=cv2.imread('image101.jpg')

rgb=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
Cgs=cv2.cvtColor(rgb,cv2.COLOR_RGB2GRAY)
```

```python
plt.imshow(rgb)
```

```python
equal_1=cv2.equalizeHist(Cgs)
```

```python
plt.subplot(2,2,1)
plt.imshow(Cgs,cmap='gray')
plt.title("Functional Conversion")

plt.subplot(2,2,2)
plt.imshow(equal_1,cmap='gray')
plt.title("Functional Histogram Equation")

plt.subplot(2,2,3)
plt.hist(Cgs.ravel(),bins=256,range=(0,255),color='red')
plt.title('Before histogram equalizer')

plt.subplot(2,2,4)
plt.hist(equal_1.ravel(),bins=256,range=(0,255),color='red')
plt.title('After histogram equalizer')


plt.tight_layout()
plt.show()
```

```python
output=cv2.imread('image102.jpg')
```

```python
def manual_histogram_equalization(gray_image):
    hist, bins = np.histogram(gray_image.flatten(), 256, [0, 256])
    cdf = hist.cumsum()
    cdf_normalized = (cdf - cdf.min()) * 255 / (cdf.max() - cdf.min())
    cdf_normalized = cdf_normalized.astype(np.uint8)
    equalized_image = cdf_normalized[gray_image]
    return equalized_image

equalized=manual_histogram_equalization(output)
```

```python
plt.subplot(2,2,1)
plt.imshow(output)
plt.title("Manual Conversion")

plt.subplot(2,2,2)
plt.imshow(equalized)
plt.title("Manual Histogram Equation")

plt.subplot(2,2,3)
plt.hist(output.ravel(),bins=256,range=(0,255),color='g')
plt.title('Before histogram equalizer')

plt.subplot(2,2,4)
plt.hist(equalized.ravel(),bins=256,range=(0,255),color='g')
plt.title('After histogram equalizer')


plt.tight_layout()
plt.show()
```

```python
resize=cv2.resize(equalized,(128,128),interpolation=cv2.INTER_CUBIC)
plt.imshow(resize)
```
