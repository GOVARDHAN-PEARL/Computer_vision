```python
from PIL import Image
import matplotlib.pyplot as plt
def extract_rgb_channels(image_path):
    image = Image.open(image_path)
    image = image.convert('RGB')
    
    # Create empty images for each channel
    r_image = Image.new('RGB', image.size)
    g_image = Image.new('RGB', image.size)
    b_image = Image.new('RGB', image.size)
    
    for y in range(image.height):
        for x in range(image.width):
            r, g, b = image.getpixel((x, y))
            
            # For red channel, set G and B to 0
            r_image.putpixel((x, y), (r, 0, 0))
            
            # For green channel, set R and B to 0
            g_image.putpixel((x, y), (0, g, 0))
            
            # For blue channel, set R and G to 0
            b_image.putpixel((x, y), (0, 0, b))

    return r_image, g_image, b_image

# Usage example:
r_image, g_image, b_image = extract_rgb_channels('image1.jpg')
r11_image, g11_image, b11_image = extract_rgb_channels('image2.jpg')

plt.subplot(2,2,1)
plt.imshow(hsv)
plt.title('Original')

plt.subplot(2,2,2)
plt.imshow(r_image)
plt.title('Red')

plt.subplot(2,2,3)
plt.imshow(g_image)
plt.title('Green')

plt.subplot(2,2,4)
plt.imshow(b_image)
plt.title('Blue')

plt.tight_layout()
plt.show()

plt.subplot(2,2,1)
plt.imshow(rgb22)
plt.title('Original')

plt.subplot(2,2,2)
plt.imshow(r11_image)
plt.title('Red')
plt.subplot(2,2,3)
plt.imshow(g11_image)
plt.title('Green')
plt.subplot(2,2,4)
plt.imshow(b11_image)
plt.title('Blue')
plt.tight_layout()
plt.show()
```

```python
def rgb2cmy(img):
    image=img.convert('RGB')
    cmy=Image.new('RGB',img.size)
    for i in range(img.height):
        for j in range(img.width):
            r,g,b=img.getpixel((j,i))
            c=255-r
            m=255-g
            y=255-b
            cmy.putpixel((j,i),(c,m,y))
    return cmy

```

```python
def extract_cmy_channels(image_path):
    image = Image.open(image_path)
    image = image.convert('CMYK')
    
    r_image = Image.new('CMYK', image.size)
    g_image = Image.new('CMYK', image.size)
    b_image = Image.new('CMYK', image.size)
    
    for y in range(image.height):
        for x in range(image.width):
            r, g, b,bl = image.getpixel((x, y))
            
    
            r_image.putpixel((x, y), (r, 0, 0,0))
            
    
            g_image.putpixel((x, y), (0, g, 0,0))
            
    
            b_image.putpixel((x, y), (0, 0, b,0))

    return r_image, g_image, b_image
```

```python
c_img,m_img,y_img=extract_cmy_channels('cmy.jpg')
c3_img,m3_img,y3_img=extract_cmy_channels('cmy2.jpg')

plt.subplot(2,2,1)
plt.imshow(cmy)
plt.title('Original')
plt.subplot(2,2,2)
plt.imshow(c_img)
plt.title('Cyan')
plt.subplot(2,2,3)
plt.imshow(m_img)
plt.title('Magenta')
plt.subplot(2,2,4)
plt.imshow(y_img)
plt.title('Yellow')

plt.tight_layout()
plt.show()

plt.subplot(2,2,1)
plt.imshow(cmy2)
plt.title('Original')
plt.subplot(2,2,2)
plt.imshow(c3_img)
plt.title('Cyan')
plt.subplot(2,2,3)
plt.imshow(m3_img)
plt.title('Magenta')
plt.subplot(2,2,4)
plt.imshow(y3_img)
plt.title('Yellow')

plt.tight_layout()
plt.show()
```

```python
def extract_hsv_channels(image_path):
    image = Image.open(image_path)
    image = image.convert('HSV')
    
    r_image = Image.new('HSV', image.size)
    g_image = Image.new('HSV', image.size)
    b_image = Image.new('HSV', image.size)
    
    for y in range(image.height):
        for x in range(image.width):
            r, g, b = image.getpixel((x, y))
            
            r_image.putpixel((x, y), (r*2,g,b))
            
            g_image.putpixel((x, y), (r, g*2, b))
            
            b_image.putpixel((x, y), (r,g, b*2))
    return r_image, g_image, b_image
```

```python
h1_img,s1_img,v1_img=extract_hsv_channels('image1.jpg')
h12_img,s12_img,v12_img=extract_hsv_channels('image2.jpg')

plt.subplot(2,2,1)
plt.imshow(hsv)
plt.title('Original')

plt.subplot(2,2,2)
plt.imshow(h1_img)
plt.title('Hue Increase by 2')

plt.subplot(2,2,3)
plt.imshow(s1_img)
plt.title('Saturation increase by 2')

plt.subplot(2,2,4)
plt.imshow(v1_img)
plt.title('Luminous increase by 2')

plt.tight_layout()
plt.show()

plt.subplot(2,2,1)
plt.imshow(hsv)
plt.title('Original')

plt.subplot(2,2,2)
plt.imshow(h12_img)
plt.title('Hue Increase by 2')

plt.subplot(2,2,3)
plt.imshow(s12_img)
plt.title('Saturation increase by 2')

plt.subplot(2,2,4)
plt.imshow(v12_img)
plt.title('Luminous increase by 2')

plt.tight_layout()
plt.show()
```

```python
def extract_hsv3_channels(image_path):
    image = Image.open(image_path)
    image = image.convert('HSV')
    
    r_image = Image.new('HSV', image.size)
    g_image = Image.new('HSV', image.size)
    b_image = Image.new('HSV', image.size)
    
    for y in range(image.height):
        for x in range(image.width):
            r, g, b = image.getpixel((x, y))
            
            r_image.putpixel((x, y), (r-100,g,b))
            
            g_image.putpixel((x, y), (r, g-100, b))
            
            b_image.putpixel((x, y), (r,g,b-100))
    return r_image, g_image, b_image
```
