```python
import numpy as np
def brin(img,fac):
    image = cv2.imread(img, cv2.IMREAD_GRAYSCALE)
    b_img=np.zeros_like(image)
    for i in range(image.shape[0]):
        for j in range(image.shape[1]):
            npq=image[i,j]+fac
            if npq>255:
                npq=255
            elif npq<0:
                npq=0
            b_img[i,j]=npq
    return b_img

    
```

```python
bri1=brin('or1.jpg',50)
```

```python
plt.subplot(2,2,1)
plt.imshow(img1)
plt.title("Original")

plt.subplot(2,2,2)
plt.imshow(bri1,cmap='gray')
plt.title("Brightness Increased")

plt.subplot(2,2,3)
plt.hist(img1.ravel(),bins=256,range=(0,255),color='red')
plt.title('Histogram plot')

plt.subplot(2,2,4)
plt.hist(bri1.ravel(),bins=256,range=(0,255),color='red')
plt.title('Histogram plot')
```

```python
def crin(img,fac):
    image = cv2.imread(img, cv2.IMREAD_GRAYSCALE)
    b_img=np.zeros_like(image)
    for i in range(image.shape[0]):
        for j in range(image.shape[1]):
            npq=fac*(image[i,j]-128)+128
            if npq>255:
                npq=255
            b_img[i,j]=npq
    return b_img
```

```python
cr1=crin('or1.jpg',1.5)
```

```python
plt.subplot(2,2,1)
plt.imshow(img1)
plt.title("Original")

plt.subplot(2,2,2)
plt.imshow(cr1,cmap='gray')
plt.title("Contrast increased")

plt.subplot(2,2,3)
plt.hist(img1.ravel(),bins=256,range=(0,255),color='red')
plt.title('Histogram plot')

plt.subplot(2,2,4)
plt.hist(cr1.ravel(),bins=256,range=(0,255),color='red')
plt.title('Histogram plot')
```

```python
def trin(img,tre):
    image = cv2.imread(img, cv2.IMREAD_GRAYSCALE)
    b_img=np.zeros_like(image)
    for i in range(image.shape[0]):
        for j in range(image.shape[1]):
            if image[i,j]>tre:
                b_img[i,j]=255
            else:
                b_img[i,j]=0
    return b_img
```

```python
def sm(image_path, kernel_size):
    img = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)
    height, width = img.shape
    sm_img = np.zeros_like(img)
    pad = kernel_size // 2
    for i in range(pad, height - pad):
        for j in range(pad, width - pad):
            kernel = img[i - pad:i + pad + 1, j - pad:j + pad + 1]
            kernel_mean = np.median(kernel)
            sm_img[i, j] = kernel_mean
    
    return sm_img
```

```python
def bl(image_path, ks):
    img = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)
    height, width = img.shape
    sm_img = np.zeros_like(img)
    kernel=np.ones((ks,ks),dtype=np.float32)/(ks*ks)
    pad = ks // 2
    for i in range(pad, height - pad):
        for j in range(pad, width - pad):
            roi = img[i - pad:i + pad + 1, j - pad:j + pad + 1]
            bv = np.sum(roi * kernel)
            sm_img[i, j] = bv
    
    return sm_img
```

```python
bl1=bl('or2.jpg',9)
```

```python
def sh(image_path):
    img = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)
    height, width = img.shape
    sm_img = np.zeros_like(img)
    kernel = np.array([[0, -1, 0],
                       [-1, 5, -1],
                       [0, -1, 0]], dtype=np.float32)
    pad = kernel.shape[0] // 2
    for i in range(pad, height - pad):
        for j in range(pad, width - pad):
            roi = img[i - pad:i + pad + 1, j - pad:j + pad + 1]
            bv = np.sum(roi * kernel)
            if bv>255:
                bv=255
            elif bv<0:
                bv=0
            sm_img[i, j] = bv
    
    return sm_img
```

```python
sh1=sh('or2.jpg')
```

```python
plt.subplot(2,2,1)
plt.imshow(img2)
plt.title("Original")

plt.subplot(2,2,2)
plt.imshow(sh1,cmap='gray')
plt.title("Sharpened image")

plt.subplot(2,2,3)
plt.hist(img2.ravel(),bins=256,range=(0,255),color='red')
plt.title('Histogram plot')

plt.subplot(2,2,4)
plt.hist(sh1.ravel(),bins=256,range=(0,255),color='red')
plt.title('Histogram plot')
```

```python
def cdft(image_path):
    img = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)
    dft = np.fft.fft2(img)
    dft_s = np.fft.fftshift(dft)
    magnitude_s = np.abs(dft_s)
    magnitude_spectrum = np.log(magnitude_s + 1)
    
    return magnitude_spectrum
```

```python
dff=cdft('or2.jpg')
```

```python
plt.subplot(2,2,1)
plt.imshow(img2)
plt.title("Original")

plt.subplot(2,2,2)
plt.imshow(dff,cmap='gray')
plt.title("Discrete Fourier transformed image")

plt.subplot(2,2,3)
plt.hist(img2.ravel(),bins=256,range=(0,255),color='red')
plt.title('Histogram plot')

plt.subplot(2,2,4)
plt.hist(dff.ravel(),bins=256,range=(0,255),color='red')
plt.title('Histogram plot')
```
